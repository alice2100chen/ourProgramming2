```
# 先產生一個list，包含40個''(空字串)，相當於一維陣列的概念
# 即 ['' for i in range(40)]

# 這樣的list共產生20個，相當於二維陣列的概念
# 即 [['' for i in range(40)] for j in range(20)]

# 最後把前述的二維結構產生3個，相當於三維陣列的概念
# 即 [[['' for i in range(40)] for j in range(20)] for k in range(3)]

name = [[['' for i in range(40)] for j in range(20)] for k in range(3)]

# 指定三位同學的姓名
name[0][0][0] = 'Alice Chen'
name[2][19][39] = 'Bob Chang'
name[0][4][31] = 'Cathy Wang'
# 逐一列印班級、座號、姓名
for i in range(3):
    for j in range(20):
        for k in range(40):
            if name[i][j][k] != '':   # 如果不是空字串(有姓名的)
                print('高', i+1 ,'第', j+1 ,'班','座號', k+1 ,'姓名:',name[i][j][k])
```
