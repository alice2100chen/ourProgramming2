```
#匯入 多工處理的套件 multiprocessing 中的 程式池函數Pool及處理器數量函數cpu_count
from multiprocessing import Pool
from multiprocessing import cpu_count
```

```
#計算平方值的函數 y = f(x) = x * x
def f(x):                
    return x * x
```

```
# cpu_count()函數會傳回目前有多少核心可用，這裡是Google Colab分配2個核心給本程式使用
cores = cpu_count()       
print ('目前有',cores,'核心')
process = Pool(cores)
num = range(3)     #num 分別為 0, 1, 2
```

```
# 方法1: imap
for y in process.imap(f, num):
    print (y)      # 依順序輸出計算結果為0, 1, 4, 
```

```
# 方法2: imap_unordered
for y in process.imap_unordered(f, num):
    print(y)      # 不必依順序輸出計算結果
```
