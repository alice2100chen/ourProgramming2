```
data = []                         #預設陣列data是空的
for i in range(6):                #連續輸入6個數字
  num = int(input('請輸入數字：'))  #讀取一個數字
  data.append(num)                #data.append()函數把變數的值「附加」到陣列data中
print('輸入的數列為',data)
```

```
data_count = len(data)
for i in range(data_count-1):
    for j in range(data_count-1-i): 
        if data[j] > data[j+1]:
            temp = data[j] 
            data[j] = data[j+1]
            data[j+1] = temp
print('排序後數列為',data)
```
